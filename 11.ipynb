```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
def head(X: list, n = 5) -> list:
    """
    return `n` (default is 5) first rows of a 2D list

    `X` : The 2D list
    `n` : number of first rows
    """
    return X[:n]

def tail(X: list, n = 5) -> list:
    """
    return `n` (default is 5) last rows of a 2D list

    `X` : The 2D list
    `n` : number of last rows
    """
    return X[(-n):]
```

```python
bcancer = datasets.load_breast_cancer()
X = bcancer.data
print("X[1] = ", head(X,1), sep='\n')
Y = bcancer.target
print("Y[1] =", head(Y, 1))
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
print(head(X,1))
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, random_state=2)
```

```python
svmc = SVC(kernel="linear")
svmc.fit(Xtrain, Ytrain)
Ypred = svmc.predict(Xtest)
svmscore = accuracy_score(Ytest, Ypred)
print("Accuracy score of Linear SVM Classifier is", round(100*svmscore, 2), "%")
```

```python
cf = confusion_matrix(Ytest, Ypred)
import seaborn as sns
sns.heatmap(data=cf, annot=True, fmt="1.1f")
plt.suptitle("Linear SVM Classifier")
```

```python
ksvmc = SVC(kernel="rbf")
ksvmc.fit(Xtrain, Ytrain)
Ypred = ksvmc.predict(Xtest)
svmcscore = accuracy_score(Ytest, Ypred)
print("Accuracy sscore of RBF kernel SVM Classifier is", round(100*svmcscore, 2), "%")
```

```python
cf = confusion_matrix(Ytest, Ypred)
import seaborn as sns
sns.heatmap(data=cf, annot=True, fmt="1.1f")
plt.suptitle("RBF SVM Classifier")
```

```python
ksvmc = SVC(kernel="poly")
ksvmc.fit(Xtrain, Ytrain)
Ypred = ksvmc.predict(Xtest)
svmcscore = accuracy_score(Ytest, Ypred)
print("Accuracy score of Poly kernel SVM Classifier is", round(100*svmcscore, 2), "%")
```

```python
cf = confusion_matrix(Ytest, Ypred)
import seaborn as sns
sns.heatmap(data=cf, annot=True, fmt="1.1f")
plt.suptitle("Poly SVM Classifier")
```
