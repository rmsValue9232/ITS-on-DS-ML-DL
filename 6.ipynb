```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
print(type(X), X.shape, type(y), y.shape)
```

```python
pca = PCA(n_components=2)
Xp = pca.fit(X).transform(X)
```

```python
plt.figure(1)
plt.scatter(Xp[:,0], Xp[:,1], c=y, cmap='jet', s=10)
plt.suptitle("Original Clusters")
plt.grid(1, which="both")
plt.axis("tight")
plt.show()
```

```python
GMM = GaussianMixture(n_components=3)
# GaussianMixture is a clustering algorithm
# TODO: learn about GaussianMixture
GMM.fit(Xp)
y_pred = GMM.predict(Xp)
plt.figure(2)
plt.scatter(Xp[:,0], Xp[:, 1], c=y_pred, cmap='jet', s=10)
plt.suptitle("GMM Clusters")
plt.grid(1, which="both")
plt.axis("tight")
plt.show()
```

```python
cmat = confusion_matrix(y, y_pred)
sns.heatmap(data=cmat, annot=True, fmt="1.1f")
```
