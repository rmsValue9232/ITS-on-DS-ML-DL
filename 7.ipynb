```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import datasets
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def head(X, n = 5):
    """
    return `n` (default is 5) first rows of a 2D list

    `X` : The 2D list
    `n` : number of first rows
    """
    return X[:n, :]

def tail(X, n = 5):
    """
    return `n` (default is 5) last rows of a 2D list

    `X` : The 2D list
    `n` : number of last rows
    """
    return X[(-n):, :]
```

```python
iris = datasets.load_iris()
X = iris.data[:50, :2]
```

```python
reg = LinearRegression().fit(X[:, 0:1], X[:, 1])
```

```python
w = reg.coef_
c = reg.intercept_
print("w =", w, ", c =", c)
print( "Mean Squared Error Loss =", mean_squared_error( X[:,1], (w[0]*X[:,0]+c) ) )
print( "r2-score =", r2_score( X[:,1], (w[0]*X[:,0]+c) ) )
```

```python
xpoints = np.linspace(4,6)
ypoints = w[0]*xpoints + c
plt.figure(1)
plt.plot(xpoints, ypoints, "g-")
plt.scatter(X[:, 0], X[:, 1], s = 10)
for i in range(0, len(X)):
    data_point = (X[i, 0], X[i, 1])
    projected_x = (data_point[0] + w[0]*(data_point[1] - c))/(1 + w[0]**2)
    # plt.scatter(projected_x, (lambda x: data_point[1] + (x-data_point[0])/(-w[0]))(projected_x), s=10)
    min_x = min(data_point[0], projected_x)
    max_x = max(data_point[0], projected_x)
    xpoints = np.linspace(min_x, max_x)
    ypoints = (lambda x: data_point[1] + (x-data_point[0])/(-w[0]))(xpoints)
    plt.plot(xpoints, ypoints, "r-.", alpha=0.2)

plt.suptitle("Linear regression on IRIS data")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.grid(1, which="both")
plt.axis("tight")
ax = plt.gca()
ax.set_aspect('equal')
plt.show()
```
