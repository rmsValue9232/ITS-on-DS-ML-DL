````python
def head(X: list, n = 5) -> list:
    """
    return `n` (default is 5) first rows of a 2D list
    ### Parameters
    `X` : The 2D list
    `n` : number of first rows
    ### Example
    ```python
    >>> X = [[1,2], [3,4], [5,6]]
    >>> print(head(X))
    [[1,2],
     [3,4],
     [5,6]]
    >>> print(head(X,2))
    [[1,2],
     [3,4]]
    ```
    """
    return X[:n]

def tail(X: list, n = 5) -> list:
    """
    return `n` (default is 5) last rows of a 2D list
    ### Parameters
    `X` : The 2D list
    `n` : number of last rows
    ### Example
    ```python
    >>> X = [[1,2], [3,4], [5,6]]
    >>> print(tail(X))
    [[1,2],
     [3,4],
     [5,6]]
    >>> print(tail(X,2))
    [[3,4],
     [5,6]]
    ```
    """
    return X[(-n):]
````

````python
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import tree
import matplotlib.pyplot as plt
````

````python
df = sns.load_dataset('iris')
X = df.iloc[:,0:4].values # selecting the columns 0,1,2,3-th
Y = df.iloc[:,4].values
print(head(X))
print(head(Y))
````

````python
cf = DecisionTreeClassifier()
# supervised learning algorithm
cf.fit(X,Y)
````

````python
# use graphviz to see the tree diagram more clearly
plt.figure(dpi=500)
decPlot = plot_tree(decision_tree=cf,feature_names=["sepal_length","sepal_width","petal_length","petal_width"],class_names=['setosa','vercicolor','verginica'],filled=True,precision=4,rounded=True,impurity=True)
````

````python
# gini is the impurity (entropy) in the current node if gini>0, it has not been correctly classified
# values represent the remainining 
text_representation = tree.export_text(cf,feature_names=["sepal_length","sepal_width","petal_length","petal_width"])
print(text_representation)
````
