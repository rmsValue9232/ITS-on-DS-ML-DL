```python
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('mobile_prices.csv')
X = df.iloc[:,:20].values
y = df.iloc[:,20:21].values
```

```python
ss = StandardScaler()
X = ss.fit_transform(X)

Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.1)
```

```python
oh = OneHotEncoder()
ytrain = oh.fit_transform(ytrain).toarray()
```

```python
model = Sequential()
model.add(Dense(16,input_dim=20,activation="relu"))
# adding the first layer of 16 neurons, with input dimensions of 20 with
model.add(Dense(12,activation="relu")) #Hidden Layer
model.add(Dense(4,activation="softmax")) #Hidden layer
# softmax used generally at the end of a multiple classification problem while sigmoid is useful only in binary classification and not in multiple classfication as the functional form of softmax include a summation of all possible incoming neurons in one 
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
# loss function is categorical_cross entropy as it is not binary classification, but multiple class classification
```

```python
history = model.fit(Xtrain,ytrain,epochs=100,batch_size=64)
```

```python
ypred = model.predict(Xtest)
ypred = np.argmax(ypred,axis=1)
ypred
```

```python
score = accuracy_score(ypred,ytest)
print('Accuracy score is',100*score,'%')
```
